```python
import os
import xml.etree.ElementTree as ET

def convert_annotation(xml_path, output_path, image_path):
    # 指定以UTF-8編碼打開文件
    with open(xml_path, 'r', encoding='utf-8') as in_file:
        tree = ET.parse(in_file)
        root = tree.getroot()
        image_name = root.find('filename').text

        size = root.find('size')
        w = int(size.find('width').text)
        h = int(size.find('height').text)

        output_file = open(os.path.join(output_path, image_name.replace('.jpg', '.txt')), 'w')

        for obj in root.findall('object'):
            difficult = obj.find('difficult').text
            if difficult == '1':
                continue
            cls = obj.find('name').text
            if cls not in classes:
                continue
            cls_id = classes.index(cls)
            xmlbox = obj.find('bndbox')
            b = (int(xmlbox.find('xmin').text), int(xmlbox.find('ymin').text), int(xmlbox.find('xmax').text), int(xmlbox.find('ymax').text))
            bb = convert((w, h), b)
            output_file.write(str(cls_id) + " " + " ".join([str(a) for a in bb]) + '\n')

        output_file.close()

def convert(size, box):
    dw = 1. / (size[0])
    dh = 1. / (size[1])
    x = (box[0] + box[2]) / 2.0 - 1
    y = (box[1] + box[3]) / 2.0 - 1
    w = box[2] - box[0]
    h = box[3] - box[1]
    x = x * dw
    w = w * dw
    y = y * dh
    h = h * dh
    return (x, y, w, h)

# 設置你的類別名稱列表
classes = ['container']  

# 設置XML文件所在目錄
xml_dir = 'yolov7-main\dataset\container\container-label\驗證集_xml'

# 設置輸出txt檔案的目錄 
output_dir = 'yolov7-main\dataset\container\container-label\驗證集'

# 設置圖片所在目錄
image_dir = 'yolov7-main\dataset\container\container-img\驗證集'

if not os.path.exists(output_dir):
    os.makedirs(output_dir)

for xml_file in os.listdir(xml_dir):
    if xml_file.endswith(".xml"):
        xml_path = os.path.join(xml_dir, xml_file)
        convert_annotation(xml_path, output_dir, image_dir)
        print(f"Converted {xml_file} to YOLO format")
```

```python
import os

def generate_image_label_list(image_dir, output_file):
    # 获取目录中的所有文件，并按照文件名排序
    files = sorted(os.listdir(image_dir))
    
    # 打开输出文件
    with open(output_file, 'w', encoding='utf-8') as outfile:
        # 遍历文件
        for file in files:
            # 检查文件是否为图片文件（假设图片文件为jpg或png格式）
            if file.lower().endswith(('.jpg', '.jpeg', '.png')):
                # 构造图片文件名和标签文件名
                image_file = os.path.join(image_dir, file)
                label_file = image_file.replace('.jpg', '.txt').replace('.jpeg', '.txt').replace('.png', '.txt')
                # 检查对应的标签文件是否存在
                if os.path.exists(label_file):
                    # 将图片文件名和标签文件名写入输出文件
                    outfile.write(image_file + "\n")
                    outfile.write(label_file + "\n")
                else:
                    print(f"Warning: Label file {label_file} not found for image {image_file}")

# 设置图片和标签文件所在目录
image_dir = 'yolov7-main/dataset/container/container-img/train'

# 设置输出的文本文件
output_file = 'yolov7-main/dataset/container/container-img/train_list.txt'

# 生成图片和标签文件列表
generate_image_label_list(image_dir, output_file)
print(f"Generated {output_file}")
```

```python
import os
from PIL import Image

# 資料夾路徑
image_folder = "dataset\container/test"
label_folder = "runs\detect\exp\labels"
output_folder = "dataset\container\cropimg"

# 確保輸出資料夾存在
os.makedirs(output_folder, exist_ok=True)

# 讀取標註文件並進行裁剪
for label_file in os.listdir(label_folder):
    if label_file.endswith(".txt"):
        # 對應的圖片文件
        image_file = os.path.splitext(label_file)[0] + ".jpg"
        image_path = os.path.join(image_folder, image_file)
        label_path = os.path.join(label_folder, label_file)
        
        if os.path.exists(image_path):
            img = Image.open(image_path)
            W, H = img.size
            
            max_conf = -1  # 初始化最大信心度
            best_label = None
            
            # 讀取標註文件
            with open(label_path, "r") as f:
                for line in f.readlines():
                    parts = line.strip().split()
                    if len(parts) == 6:
                        cls, x, y, w, h, conf = map(float, parts)
                        
                        # 找到信心度最高的標註
                        if conf > max_conf:
                            max_conf = conf
                            best_label = (cls, x, y, w, h, conf)
            
            # 使用信心度最高的標註進行裁剪
            if best_label is not None:
                cls, x, y, w, h, conf = best_label

                # 將歸一化座標轉為像素座標
                center_x = int(x * W)
                center_y = int(y * H)
                width = int(w * W)
                height = int(h * H)

                # 計算左上角和右下角座標
                left = max(center_x - width // 2, 0)
                top = max(center_y - height // 2, 0)
                right = min(center_x + width // 2, W)
                bottom = min(center_y + height // 2, H)

                # 裁切並保存圖片
                cropped = img.crop((left, top, right, bottom))
                cropped_file_name = f"{os.path.splitext(image_file)[0]}_cropped_{cls}_{int(conf*100)}.jpg"
                cropped.save(os.path.join(output_folder, cropped_file_name))

print("裁剪完成。")
```

```python
from PIL import Image, ImageFilter, ImageEnhance
import re
import easyocr
import numpy as np
import os

# 指定要遍歷的資料夾路徑
folder_path = 'dataset\container\cropimg_test'

# 初始化EasyOCR reader
reader = easyocr.Reader(['en'])

# 計數器
total_count = 0
correct_count = 0

# 遍歷資料夾中的所有文件
for filename in os.listdir(folder_path):
    # 構建文件完整路徑
    file_path = os.path.join(folder_path, filename)
    
    # 檢查是否為圖片文件
    if os.path.isfile(file_path) and (file_path.endswith('.jpg') or file_path.endswith('.png')):
        total_count += 1  # 統計總數
        
        # 獲取文件名稱(不包含副檔名)
        file_name = os.path.splitext(filename)[0]
        
        # 打開圖片
        img = Image.open(file_path)
        
        # 預處理
        img = img.convert('L')
        enhancer = ImageEnhance.Contrast(img)
        contrast_image = enhancer.enhance(2) #對比度
        sharp_image = contrast_image.filter(ImageFilter.SHARPEN) #銳利化
        
        # PIL轉NUMPY
        img_np = np.array(sharp_image)
        
        # 執行OCR
        result = reader.readtext(img_np, detail=0)
        s = ','.join(result)
        new_string = re.sub(r"[^a-zA-Z0-9]", "", s)[:10]
        
        # 檢查識別結果是否與文件名相同
        if new_string == file_name[:10]:
            correct_count += 1
            result_str = "成功"
        else:
            result_str = "失敗"
        
        # 輸出結果
        print(f"File: {filename}, Text: {new_string} ({result_str})")

# 計算準確率
accuracy = correct_count / total_count if total_count > 0 else 0
print(f"準確率: {accuracy:.2f}")
```
